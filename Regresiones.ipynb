```python
import pandas as pd

# Crear el diccionario con los datos
data = {
    'Representantes de ventas': [
        'Brian Virost', 'Carlos Ramirez', 'Carol Sala', 'Greg Fish',
        'Jeff Hall', 'Mark Reynolds', 'Meryl Rumsey', 'Mike Kiel',
        'Ray Snarsky', 'Rich Niles', 'Ron Broderick', 'Sal Spina',
        'Soni Jones', 'Susan Welch', 'Tom Keller'
    ],
    'Número de llamadas de ventas': [
        96, 40, 104, 128, 164, 76, 72, 80,
        36, 84, 180, 132, 120, 44, 84
    ],
    'Número de copiadoras vendidas': [
        41, 41, 51, 60, 61, 29, 39, 50,
        28, 43, 70, 56, 45, 31, 30
    ]
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

# Histograma para 'Número de llamadas de ventas'
plt.figure(figsize=(8, 6))
plt.hist(df['Número de llamadas de ventas'], bins=10, edgecolor='black')
plt.title('Distribución del Número de Llamadas de Ventas')
plt.xlabel('Número de llamadas')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Histograma para 'Número de copiadoras vendidas'
plt.figure(figsize=(8, 6))
plt.hist(df['Número de copiadoras vendidas'], bins=10, edgecolor='black', color='orange')
plt.title('Distribución del Número de Copiadoras Vendidas')
plt.xlabel('Número de copiadoras')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import shapiro

# Realizar el test de Shapiro-Wilk para 'Número de llamadas de ventas'
shapiro_test_llamadas = shapiro(df['Número de llamadas de ventas'])
print(f'Test de Shapiro-Wilk para Número de llamadas de ventas: Estadística={shapiro_test_llamadas.statistic:.3f}, p-valor={shapiro_test_llamadas.pvalue:.3f}')

# Realizar el test de Shapiro-Wilk para 'Número de copiadoras vendidas'
shapiro_test_copiadoras = shapiro(df['Número de copiadoras vendidas'])
print(f'Test de Shapiro-Wilk para Número de copiadoras vendidas: Estadística={shapiro_test_copiadoras.statistic:.3f}, p-valor={shapiro_test_copiadoras.pvalue:.3f}')

# Interpretar los resultados y determinar el coeficiente más representativo
alpha = 0.05  # Nivel de significancia

print("\nInterpretación:")
if shapiro_test_llamadas.pvalue > alpha and shapiro_test_copiadoras.pvalue > alpha:
    print("Ambas variables parecen seguir una distribución normal (p-valor > 0.05). El coeficiente de Pearson es el más representativo.")
elif shapiro_test_llamadas.pvalue <= alpha and shapiro_test_copiadoras.pvalue <= alpha:
    print("Ninguna de las variables parece seguir una distribución normal (p-valor <= 0.05). Los coeficientes de Kendall o Spearman son más representativos.")
elif shapiro_test_llamadas.pvalue <= alpha:
    print("La variable 'Número de llamadas de ventas' no parece seguir una distribución normal (p-valor <= 0.05). Los coeficientes de Kendall o Spearman son más representativos.")
else: # shapiro_test_copiadoras.pvalue <= alpha
    print("La variable 'Número de copiadoras vendidas' no parece seguir una distribución normal (p-valor <= 0.05). Los coeficientes de Kendall o Spearman son más representativos.")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro
import pandas as pd

# 1. Visualización de Distribuciones (Histogramas)
print("Generando histogramas:")
plt.figure(figsize=(8, 6))
plt.hist(df['Número de llamadas de ventas'], bins=10, edgecolor='black')
plt.title('Distribución del Número de Llamadas de Ventas')
plt.xlabel('Número de llamadas')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(df['Número de copiadoras vendidas'], bins=10, edgecolor='black', color='orange')
plt.title('Distribución del Número de Copiadoras Vendidas')
plt.xlabel('Número de copiadoras')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# 2. Visualización de la Relación (Diagrama de Dispersión)
print("\nGenerando diagrama de dispersión:")
plt.figure(figsize=(8, 6))
plt.scatter(df['Número de llamadas de ventas'], df['Número de copiadoras vendidas'])
plt.title('Relación entre Número de Llamadas y Copiadoras Vendidas')
plt.xlabel('Número de llamadas de ventas')
plt.ylabel('Número de copiadoras vendidas')
plt.grid(True)
plt.show()

# 3. Test de Normalidad (Shapiro-Wilk)
print("\nRealizando test de normalidad (Shapiro-Wilk):")
shapiro_test_llamadas = shapiro(df['Número de llamadas de ventas'])
print(f'Test de Shapiro-Wilk para Número de llamadas de ventas: Estadística={shapiro_test_llamadas.statistic:.3f}, p-valor={shapiro_test_llamadas.pvalue:.3f}')

shapiro_test_copiadoras = shapiro(df['Número de copiadoras vendidas'])
print(f'Test de Shapiro-Wilk para Número de copiadoras vendidas: Estadística={shapiro_test_copiadoras.statistic:.3f}, p-valor={shapiro_test_copiadoras.pvalue:.3f}')

# 4. Cálculo de Coeficientes de Correlación
print("\nCalculando coeficientes de correlación:")
pearson_corr = df['Número de llamadas de ventas'].corr(df['Número de copiadoras vendidas'], method='pearson')
kendall_corr = df['Número de llamadas de ventas'].corr(df['Número de copiadoras vendidas'], method='kendall')
spearman_corr = df['Número de llamadas de ventas'].corr(df['Número de copiadoras vendidas'], method='spearman')

print(f'Coeficiente de correlación de Pearson: {pearson_corr:.2f}')
print(f'Coeficiente de correlación de Kendall: {kendall_corr:.2f}')
print(f'Coeficiente de correlación de Spearman: {spearman_corr:.2f}')

# 5. Determinación del Coeficiente más Representativo
print("\nDeterminando el coeficiente más representativo:")
alpha = 0.05  # Nivel de significancia

if shapiro_test_llamadas.pvalue > alpha and shapiro_test_copiadoras.pvalue > alpha:
    print("Ambas variables parecen seguir una distribución normal (p-valor > 0.05). El coeficiente de Pearson es el más representativo.")
elif shapiro_test_llamadas.pvalue <= alpha and shapiro_test_copiadoras.pvalue <= alpha:
    print("Ninguna de las variables parece seguir una distribución normal (p-valor <= 0.05). Los coeficientes de Kendall o Spearman son más representativos.")
elif shapiro_test_llamadas.pvalue <= alpha:
    print("La variable 'Número de llamadas de ventas' no parece seguir una distribución normal (p-valor <= 0.05). Los coeficientes de Kendall o Spearman son más representativos.")
else: # shapiro_test_copiadoras.pvalue <= alpha
    print("La variable 'Número de copiadoras vendidas' no parece seguir una distribución normal (p-valor <= 0.05). Los coeficientes de Kendall o Spearman son más representativos.")
```
