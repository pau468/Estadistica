```python
import pandas as pd

# Load the dataframe
df = pd.read_csv("DatosNacimientos.csv", encoding='latin-1')

# Define the list of filtered columns
df_filtrado_columnas=["PESO (Gramos)" , "HORA NACIMIENTO" , "CODIGO INSTITUCION" , "DEPARTAMENTO" , "AREA NACIMIENTO" , "GRUPO SANGUINEO"]

# Identify numerical columns from the filtered list
numerical_cols = [col for col in df_filtrado_columnas if df[col].dtype in ['float64', 'int64']]

df_cleaned = df.copy()

for col in numerical_cols:
    print(f"\nProcessing column: {col}")

    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    initial_rows = len(df_cleaned)
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
    removed_rows = initial_rows - len(df_cleaned)

    print(f"Removed {removed_rows} outliers from column '{col}'")

print("\nShape of the original DataFrame:", df.shape)
print("Shape of the DataFrame after removing outliers:", df_cleaned.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns from the filtered list
numerical_cols = [col for col in df_filtrado_columnas if col in df_cleaned.columns and df_cleaned[col].dtype in ['float64', 'int64']]

for col in numerical_cols:
    print(f"\nGenerating histograms for cleaned column: {col}")

    # Histogram
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df_cleaned, x=col, kde=True, palette='viridis')
    plt.title(f'Distribution of {col} (after outlier removal)')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm

# Load the dataframe
df = pd.read_csv("DatosNacimientos.csv", encoding='latin-1')

# Define the list of filtered columns (assuming these are the ones the user is interested in)
df_filtrado_columnas=["PESO (Gramos)" , "HORA NACIMIENTO" , "CODIGO INSTITUCION" , "DEPARTAMENTO" , "AREA NACIMIENTO" , "GRUPO SANGUINEO"]

# Identify numerical columns from the filtered list
numerical_cols = [col for col in df_filtrado_columnas if df[col].dtype in ['float64', 'int64']]

# Assuming continuous numerical columns are a subset of numerical_cols
# For this dataset, let's assume 'PESO (Gramos)' is a continuous numerical variable.
# 'CODIGO INSTITUCION' is likely categorical or discrete, so we'll exclude it from normality tests for continuous variables.
continuous_numerical_cols = ['PESO (Gramos)']


for col in continuous_numerical_cols:
    print(f"\nPerforming normality tests for column: {col}")

    # Remove NaN values for normality tests
    data = df[col].dropna()

    if len(data) > 5000:
        print("Kolmogorov-Smirnov Test:")
        ks_statistic, p_value_ks = stats.kstest(data, 'norm')
        print(f"  Statistic: {ks_statistic:.4f}, P-value: {p_value_ks:.4f}")
    else:
        print("Shapiro-Wilk Test:")
        shapiro_statistic, p_value_shapiro = stats.shapiro(data)
        print(f"  Statistic: {shapiro_statistic:.4f}, P-value: {p_value_shapiro:.4f}")

    print("\nAnderson-Darling Test:")
    anderson_result = stats.anderson(data, dist='norm')
    print(f"  Statistic: {anderson_result.statistic:.4f}")
    for i in range(len(anderson_result.critical_values)):
        sl, cv = anderson_result.significance_level[i], anderson_result.critical_values[i]
        print(f"  {sl}% Critical Value: {cv:.4f}")

    print("\nJarque-Bera Test:")
    # Jarque-Bera test requires at least 2000 data points
    if len(data) >= 2000:
        jb_statistic, p_value_jb = stats.jarque_bera(data)
        print(f"  Statistic: {jb_statistic:.4f}, P-value: {p_value_jb:.4f}")
    else:
        print("  Jarque-Bera test requires at least 2000 data points.")


    print(f"\nGenerating QQ plot for column: {col}")
    plt.figure(figsize=(8, 6))
    sm.qqplot(data, line='s')
    plt.title(f'QQ Plot of {col}')
    plt.show()
```
